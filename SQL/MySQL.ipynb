```python
# db 연결을 활성화해주는 함수 구현

def conn(d_name):
    import pymysql
    host_name = 'localhost'
    host_port = 3306
    username = 'root'
    password = 'jimkwon95'
    database_name = d_name
    db = pymysql.connect(
        host=host_name,
        port=host_port,
        user=username,
        passwd=password,
        db=database_name,
        charset='utf8'
    )
    return db
```

```python
db = conn('sqlDB')
```

```python
cursor = db.cursor()
sql_query = "INSERT INTO userTbl VALUES('STJ','서태지',1975,'경기','011','00000000',171,'2014-05-01')"
cursor.execute(sql_query)
db.commit()
```

```python
sql_query = "INSERT INTO buyTbl (userID,prodName,groupName,price,amount) VALUES('STJ','운동화',NULL,50,5)"
cursor.execute(sql_query)
db.commit()
```
