```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rc("font", family="Malgun Gothic") #이거 실행해줘야 한글이 주피터에서 깨지지 않는다.
```

```python
df1 = pd.read_excel('notExercise.xlsx')
df1.head()
```

```python
df1.drop(columns='기간',inplace=True)
df1.tail()
```

```python
df1['대분류'][23:]
```

```python
df1.drop(index=range(23,53),inplace=True)
df1
```

```python
df_sex=df1[df1['대분류']=='성별'].copy()
df_sex
```

```python
df_sex.drop(columns='대분류',inplace=True)
df_sex.set_index('분류',inplace=True)

df_sex
```

```python
figure,ax=plt.subplots(2,2,figsize=(10,10))

df_sex['운동을 할 충분한 시간이 없어서'].plot.pie(explode=[0,0.02],
                                                ax = ax[0][0],
                                                autopct = '%1.1f%%')
ax[0][0].set_title('운동을 할 충분한 시간이 없어서')
ax[0][0].set_ylabel('')

df_sex['함께 운동을 할 사람이 없어서'].plot.pie(explode=[0,0.02],
                                                ax = ax[0][1],
                                                autopct = '%1.1f%%')
ax[0][0].set_title('함께 운동을 할 사람이 없어서')
ax[0][0].set_ylabel('')

df_sex['운동을 싫어해서'].plot.pie(explode=[0,0.02],
                                                ax = ax[1][0],
                                                autopct = '%1.1f%%')
ax[0][0].set_title('운동을 싫어해서')
ax[0][0].set_ylabel('')

df_sex['운동을 할 만한 장소가 없어서'].plot.pie(explode=[0,0.02],
                                                ax = ax[1][1],
                                                autopct = '%1.1f%%')
ax[0][0].set_title('운동을 할 만한 장소가 없어서')
ax[0][0].set_ylabel('')

plt.show()
```

```python
df_age=df1[df1['대분류']=='연령별'].copy()
df_age
```

```python
df_age.drop(columns='대분류',inplace=True)
df_age.set_index('분류',inplace=True)

df_age
```

```python
figure,ax=plt.subplots(1,3,figsize=(16,8))

explode_set = [0.02,0.02,0.02,0.02,0.02,0.02]

df_age['운동을 할 충분한 시간이 없어서'].plot.pie(explode=explode_set,
                                                ax = ax[0],
                                                autopct = '%1.1f%%')
ax[0].set_title('운동을 할 충분한 시간이 없어서')
ax[0].set_ylabel('')

df_age['함께 운동을 할 사람이 없어서'].plot.pie(explode=explode_set,
                                                ax = ax[1],
                                                autopct = '%1.1f%%')
ax[1].set_title('함께 운동을 할 사람이 없어서')
ax[1].set_ylabel('')

df_age['운동을 싫어해서'].plot.pie(explode=explode_set,
                                                ax = ax[2],
                                                autopct = '%1.1f%%')
ax[2].set_title('운동을 싫어해서')
ax[2].set_ylabel('')

plt.show()
```
